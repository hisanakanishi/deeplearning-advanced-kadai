```python
from bs4 import BeautifulSoup
from urllib import request
```

```python
url = 'https://www.aozora.gr.jp/cards/000148/files/2371_13943.html'
response = request.urlopen(url)
soup = BeautifulSoup(response)
response.close()
main_text = soup.find('div', class_='main_text')
```

```python
import re
import MeCab
main_text = main_text.get_text()
text2 =re.sub('\n|\u3000|\r', "", main_text)
```

```python
mecab_tagger = MeCab.Tagger()
parsed_txt = mecab_tagger.parse(text2)
elements = parsed_txt.split('\n')
```

```python
results = []

for element in elements:
    parts = element.split('\t')[0]
    results.append(parts)
```

```python
#辞書の取得
url_dic = "http://svn.sourceforge.jp/svnroot/slothlib/CSharp/Version1/SlothLib/NLP/Filter/StopWord/word/Japanese.txt"
re2 = request.urlopen(url_dic)
soup2 = BeautifulSoup(re2)
re2.close()
stopwords_text = soup2.text
stopwords_list = stopwords_text.split("\r\n")
stopwords_list = [word for word in stopwords_list if word]
```

```python
result_text_list = list()
for result in results:
    if result not in stopwords_list:
        result_text_list.append(result)
result_text_list = result_text_list[:-2]        
print(result_text_list)
```
